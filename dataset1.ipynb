```python
from datasets import load_dataset
from google.colab import userdata
from huggingface_hub import login
import pandas as pd

login(token=userdata.get("HF_TOKEN"))

DATASET_NAME = "rogue-security/prompt-injections-benchmark"
TARGET_LABEL = "jailbreak"
SAMPLE_SIZE  = 100
RANDOM_SEED  = 42  # Fixed for reproducibility
```

```python
dataset = load_dataset(DATASET_NAME, split="test")
df = dataset.to_pandas()

df_jailbreak = df[df["label"] == TARGET_LABEL].reset_index(drop=True)
df_sample = df_jailbreak.sample(n=SAMPLE_SIZE, random_state=RANDOM_SEED).reset_index(drop=True)

print(f"Dataset: {DATASET_NAME}")
print(f"Total rows: {len(df)} | Label='{TARGET_LABEL}': {len(df_jailbreak)} | Sampled: {len(df_sample)}")
df_sample.head()
```

```python
from google.colab import files

OUTPUT_FILE = f"jailbreak_sample_{SAMPLE_SIZE}_seed{RANDOM_SEED}.csv"
df_sample.to_csv(OUTPUT_FILE, index=False)
files.download(OUTPUT_FILE)
```
