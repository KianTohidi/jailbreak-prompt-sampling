```python
from datasets import load_dataset
from google.colab import userdata
from huggingface_hub import login
import pandas as pd

login(token=userdata.get("HF_TOKEN"))

DATASET_NAME = "allenai/wildjailbreak"
SAMPLE_SIZE  = 100
RANDOM_SEED  = 42  # Fixed for reproducibility
```

```python
dataset = load_dataset(DATASET_NAME, "train", delimiter="\t", keep_default_na=False)
df = dataset["train"].to_pandas()

print(f"Dataset: {DATASET_NAME}")
print(f"Total rows: {len(df)}")
print(f"Columns available: {list(df.columns)}")
print(f"\nLabel distribution:\n{df['data_type'].value_counts()}\n")

# Filter for adversarial harmful only (true jailbreak attempts)
df_jailbreak = df[df["data_type"] == "adversarial_harmful"].reset_index(drop=True)
print(f"Rows with label 'adversarial_harmful': {len(df_jailbreak)}")

df_sample = df_jailbreak.sample(n=SAMPLE_SIZE, random_state=RANDOM_SEED).reset_index(drop=True)

# Keep only relevant columns and rename for consistency with first dataset
df_sample = df_sample[["adversarial", "data_type"]].rename(columns={"adversarial": "text", "data_type": "label"})

print(f"Sampled: {len(df_sample)} rows with seed={RANDOM_SEED} ✓")
df_sample.head()
```

```python
from google.colab import files

OUTPUT_FILE = f"wildjailbreak_sample_{SAMPLE_SIZE}_seed{RANDOM_SEED}.csv"
df_sample.to_csv(OUTPUT_FILE, index=False)
files.download(OUTPUT_FILE)
```
